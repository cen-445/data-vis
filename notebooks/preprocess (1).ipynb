```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#path
RAW_DATA_PATH = r"..\data\raw\Telco-Customer-Churn.csv"
PROCESSED_DATA_PATH = r"..\data\processed\Telco_processed.csv"
```

```python
df_raw = pd.read_csv(RAW_DATA_PATH)
print(df_raw.head(3))
```

```python
print("Raw shape: " ,df_raw.shape)
print(df_raw.info())
```

```python
df = df_raw.drop(['customerID'], axis = 1) # anlam ifade etmediği için customerID sildik
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

```python
df = df.fillna(df["TotalCharges"].mean())
df.isnull().sum()
```

```python
print(df[df['tenure'] == 0].index)
df.drop(labels=df[df['tenure'] == 0].index, axis=0, inplace=True)
```

```python
df["SeniorCitizen"]= df["SeniorCitizen"].map({0: "No", 1: "Yes"})
```

```python
cat_cols = df.select_dtypes(include="object").columns.tolist()
num_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()

print("Categorical:", cat_cols)
print("Numerical:", num_cols)
```

```python
label_encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
X = df.drop(columns=["Churn"])
y = df["Churn"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)
```

```python
df_processed = df.copy()
df_processed.to_csv(PROCESSED_DATA_PATH, index=False)
```
